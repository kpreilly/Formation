## Practice Problem: Determine if employee is a direct report

Given an org chart of a company, employee A and employee B, figure out if employee B is a direct report of employee A

````python
# Needed classes
class Node:
    def __init__(self, value, left=None, right=None) -> None:
        self.value = value
        self.left =left
        self.right = right
````

````python
'''
- Find the node whose value matches manager
- Starting with manager as the root, search it's children
- if the values of left or right child match employee, return True
- else Keep searching
'''

def isDirectReport(person: Node, manager: int, employee: int):
    if person is None:
        return False
    if person.value == manager:
        left = False
        right = False
        if person.left:
            left = person.left.value == employee
        if person.right:
            right = person.right.value == employee
        return left or right
    isDirectReport(person.left, manager, employee)
    isDirectReport(person.right, manager, employee)
    
````

## Pre, In, and Post order traversal

**Pre-Order**

```text
preOrder(t):
    if (t is not empty) {
        process t's root element
        preOrder( left child )
        preOrder( right child )
    }
```

**In-Order**

```text
inOrder(t) {
    if (t is not empty) {
        inOrder( left child )
        process t
        inOrder( right child )
    }
}
```

**Post-Order**

```text
postOrder(t) {
    if (t is not empty) {
        postOrder( left child )
        postOrder( right child )
        process t.value
    }
}
```

## Recursion assessment


Given a set of units and a target value, return all of the ways that these units may be combined to achieve the target value. For example, for units `[1, 2]` and a target of 4, the output should be:

```text
[
  [1, 1, 1, 1],
  [1, 1, 2],
  [2, 2]
]
```

Notice that `[1, 2, 1]` and `[2, 1, 1]` are also possibilities but we only return the one where the values are sorted from smallest to largest.

The units will always be a non-empty array of positive integers and sorted from small to large. The output should be ordered by those using the small units first before moving to larger and larger units.

````python
def solution(units, target):
    result = []
    helper2(units, target, [], result)
    return result

def helper(units, target, result, current, idx):
    if sum(current) == target:
        _ = sorted(current)
        if _ not in result:
            result.append(_)
        while sum(current) + units[idx] > target:
            current.pop()
    if idx >= len(units):
        return
    while sum(current) + units[idx] <= target:
        current.append(units[idx])
    if sum(current) != target:
        result = []
        return

    helper(units, target, result, current, idx)


def helper2(units, target, path, ret):
    if target < 0:
        return
    if target == 0:
        ret.append(path)
        return
    for i in range(len(units)):
        helper2(units[i:], target-units[i], path+[units[i]], ret)

'''
solution([1,2], 4) should result in the following matrix
[
    [1, 1, 1, 1],
    [1, 1, 2],
    [2, 2]
]
'''

print(solution([1,2], 4))
````

## Practice Problem: Flatten nested list sum

**Prompt**

Given a nested array where each element may be 1) an integer or 2) an array, whose elements themselves may be integers or further arrays, compute the sum of all the integers in the nested array.

What is the shape or pattern of this nested array structure?

As a follow on, modify this code to multiply each list sum by it's depth in the nesting. `[1, 2]` returns 3 since $(1 + 2)$ is only inside of one array. But `[1, [2, 3]]` now evaluates to 11 because the $(2 + 3)$ is at depth 2, so $5 * 2$ is 10, then add the one for 11.

````python
from typing import List


def sumNestedList(nestedList: List) -> int:
    accumulator = 0
    i = 0
    while i < len(nestedList):
        if isinstance(nestedList[i], int):
            accumulator += nestedList[i]
            i += 1
        else:
            # Flatten the list
            nestedList += nestedList[i][:]
            nestedList.pop(i)
    return  accumulator

print(sumNestedList([1, 2, 3]) == 6, 'should be True')
print(sumNestedList([1, [1, 2, 3], 3]) == 10, 'should be True')
print(sumNestedList([1, [1, [1, [1, [1]]]]]) == 5, 'should be True')
````

Explanation:

We want to keep track of the sum so an accumulator will be required.

For every element in the list we can either:

1. Add it to the accumulator if it's an int
2. Flatten it if it's a list.

Time complexity: O(n) where n is the number of integers in nestedList + O(l) where l is the number of lists we must flatten.

````python
"""


Notes:
- NxN matrix

Function Signature
func isToeplitz(input)


How do we traverse a diagonal from a starting point?
- Add 1 to row and column until we reach the end of matrix
- Key == First Element in a Diagonal

Approach:
- locate bottom left index
- go up and to the right to go through the entire matrix
- check all the diagonals by increment col by 1
- check if values are not equal return false
- return true
"""

from typing import List
"""
1 3 8 2
2 1 3 8
8 2 1 3
3 8 2 1
"""
def is_toeplitz(matrix: List[List[int]]) -> bool:
    # locate bottom left index
    # for(let row = matrix.length; row > -1; row--)
    for row in range(len(matrix)-1, -1, -1): # N - 1 to 0
        # check all the diagonals by increment col by 1
        if not check_diagonal(matrix, row, 0):
            return False

    # TODO: Check diagonals for first row
    for col in range(len(matrix)):
        if not check_diagonal(matrix, 0, col):
            # check if values are not equal return false
            return False
            
    # return true
    return True

# TODO: Arguments
def check_diagonal(matrix, row, col):
    key = matrix[row][col]
    while col < len(matrix) and row < len(matrix):
        if matrix[row][col] != key:
            return False
        col += 1
        row += 1
        
    return True




````
